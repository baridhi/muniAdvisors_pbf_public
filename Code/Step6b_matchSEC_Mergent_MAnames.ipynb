```python
import openai
import time
```

```python
# Set API key
openai_api_key ='DanGarrett_WhartonAPIkey'
client = openai.OpenAI(api_key=openai_api_key)
```

```python
# Example lists (replace with our actual data)
db1_names = ["Smith Advisors LLC", "Taylor & Co", "Johnson Municipal Advisory Services"]
db2_names = ["Smith Advisory L.L.C.", "Tayler and Company", "Jonson Municipal Advisors", "Global Capital"]
```

```python
def get_best_match(query_name, candidate_names, model="gpt-3.5-turbo", temperature=0):
    prompt = f"""You are a careful data analyst matching municipal advisor firm names across two databases. 
Given the firm name from Database A: "{query_name}"

Which of the following names from Database B is the best match? If none match closely, say "No good match". 

List of candidate names:
{chr(10).join(f"- {name}" for name in candidate_names)}

Respond with just the best matching name, or "No good match".
"""

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a meticulous data analyst skilled in entity resolution."},
            {"role": "user", "content": "Which firm matches..."},  # your prompt here
        ],
        temperature=0,
    )
    print(response.choices[0].message.content)


    return response['choices'][0]['message']['content'].strip()
```

```python
# Perform matching
matches = []
for name in db1_names:
    match = get_best_match(name, db2_names)
    print(f'"{name}" matched with: {match}')
    matches.append((name, match))
    time.sleep(1)  # Respect API rate limits

# Optional: Write results to CSV
import csv
with open(r'C:\Users\barid\Documents\Papers\Muni_adv_PBF\Data\Step6b_matched_names.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['DB1 Name', 'Matched DB2 Name'])
    writer.writerows(matches)
```
